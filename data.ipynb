```python
import os
import json
import copy
```

```python
res_path="RPT/tools/results/VOT2019/siamreppoints/baseline/"
to_path="/home/hw6/predict"
```

```python
v_dict={}

with open('RPT/testing_dataset/VOT2019/VOT2019.json','r') as d:
    dt=json.load(d)
    for video in dt:
        detail=dt[video]
        v_dict[video]=detail["init_rect"]

for video in os.listdir(res_path):
    video_path=os.path.join(res_path,video)
    pred=[v_dict[video]]
    left=None
    right=None
    flag=False
    for f in os.listdir(video_path):
#         print(f)
        if not f.endswith(".txt") or "checkpoint" in f:
            continue
        to_name=f.replace("_001","")
        f_path=os.path.join(video_path,f)
        init=to_name.replace(".txt","")
        with open(f_path,'r') as from_file:
            
            lines=from_file.readlines()
            cnt=0
#             print(lines)
            for l in lines[1:]:
                cnt+=1
                pos=[float(x) for x in l.strip().split(",")]
#                 print(pos)
                if len(pos)!=4:
                    if not flag:
                        left=cnt-1
                        flag=True
                else:
                    cor=[pos[0],pos[1],pos[0]+pos[2],pos[1],pos[0]+pos[2],pos[1]+pos[3],pos[0],pos[0]+pos[3]]
                    if flag:
                        right=cnt
                        flag=False
                        dis=[]
                        interval=right-left
                        first=[]
                        for i in range(8):
                            print(cor,pred[left])
                            dis.append((cor[i]-pred[left][i])/interval)
                            first.append(pred[left][i]+dis[i])
                        pred.append(copy.deepcopy(first))
                        cur=copy.deepcopy(first)
                        for i in range(left+2,right):
                            for j in range(8):
                                cur[j]=cur[j]+dis[j]
                            pred.append(copy.deepcopy(cur))
                    pred.append(cor)
                    
        to_file=os.path.join(to_path,to_name)
        with open(to_file,'w') as tf:
            for p in pred:
                tf.write(",".join([str(x) for x in p])+"\n")
                            
        
```
